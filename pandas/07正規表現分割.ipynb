```python
import pandas as pd
```

```python
df = pd.read_csv("head.csv")
```

```python
df.head()
```

```python
df.drop(columns=["Unnamed: 12"], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["年月日"] = pd.to_datetime(df["年月日"], format='%Y年%m月%d日')
```

```python
df.head()
```

```python
# 名前付きグループ カラム名
df2 = df["競馬場"].str.extract('(?P<回数>\d*?)回(?P<場所>.*?)(?P<日程>\d.*)日', expand=True)
```

```python
df2.head()
```

```python
df = pd.concat([df, df2], axis=1)
```

```python
df.head()
```

```python
# 条件
df["条件"] = df["条件"].str.extract("(.*?)(\(|\[|$)")[0]
```

```python
df["条件"].value_counts()
```

```python
df["条件"] = df["条件"].str.replace("牡・牝", "")
```

```python
df["条件"] = df["条件"].str.replace("牝", "")
```
