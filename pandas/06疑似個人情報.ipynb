```python
import pandas as pd
```

```python
df = pd.read_pickle("05疑似個人情報.pkl")
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
# 生年月日をto_datetime()
df["生年月日"] = pd.to_datetime(df["生年月日"])
```

```python
df.info()
```

```python
# 1985年以降生まれの女性
df.loc[(df["性別"] == "女") & (df["生年月日"].dt.year >= 1985), :]
# df["生年月日"] >= "1985" 文字列でもいける
```

```python
# 4月生まれ女性
df.loc[(df["性別"] == "女") & (df["生年月日"].dt.month == 4), :]
```

```python
# A型の男性
df.loc[(df["性別"] == "男") & (df["血液型"] == "A"), :]
```

```python
# 出身地 北海道をほっかいどうに置換
df.loc[df["出身地"] == "北海道", "出身地"] = "ほっかいどう"
df
```

```python
# 出身地の種類
df["出身地"].value_counts()
```

```python
# 血液型の割合
df["血液型"].value_counts(normalize=True)
```

```python
# 男女比
df["性別"].value_counts(normalize=True)
```

```python
# 年齢の平均
df["年齢"].mean()
```

```python
# 年齢の最小
df["年齢"].min()
```

```python
# 年齢の最大
df["年齢"].max()
```

```python
# 年代カラムを作成
df["年代"] = pd.cut(df["年齢"], [0, 10, 20, 30, 40 ,50 ,60])
df.head()
```

```python
# 年代、血液型をグループ化
g = df.groupby(["年代","血液型"])
```

```python
g["血液型"].count()
```

```python
# 性別と血液型をダミー変数
df = pd.get_dummies(df, columns=["性別", "血液型"])
df
```

```python
# カタカナにアがつく人はカラム「ア」に1
df["ア"] = 0
```

```python
df.loc[df["氏名（カタカナ）"].str.contains("ア"), "ア"] = 1
```
