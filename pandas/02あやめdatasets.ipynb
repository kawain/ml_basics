```python
import pandas as pd
from sklearn import datasets
```

```python
iris = datasets.load_iris()
print(type(iris.data))
print(iris.data)
print("-----")
print(type(iris.target))
print(iris.target)
print("-----")
print(type(iris.feature_names))
print(iris.feature_names)
```

```python
# numpy配列をDataFrameにして見やすくする
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
# カラム追加
df["target"] = iris.target
```

```python
df
```

```python
# csvに書き込み
df.to_csv("iris.csv")
```

```python
# csv読込
iris_csv = pd.read_csv("iris.csv")
iris_csv
```

```python
# csvに書き込み index=False にしてUnnamed: 0を出さない
df.to_csv("iris.csv", index=False)
```

```python
# csv読込
iris_csv = pd.read_csv("iris.csv")
iris_csv
```

## pandas を使っているならば、import pickle は不要

DataFrameでも、それ以外のオブジェクトでも保存可能な pd.to_pickle

```python
# df.to_pickle("保存ファイル名") これで書かれる例が多いが
# pd.to_pickle(オブジェクト名, "保存ファイル名") ← この書き方に統一した方がいい
pd.to_pickle(df, "iris.pkl")
```

```python
# 理由
# これはエラー
a = [1,2,3]
a.to_pickle("a.pkl")
```

```python
# これは成功
pd.to_pickle(a, "a.pkl")
```

```python
# 読み出し
hoge = pd.read_pickle("a.pkl")
hoge
```

```python
# 読み出し
iris_df = pd.read_pickle("iris.pkl")
iris_df
```
